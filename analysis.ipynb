📊 Key Business Insights from EDA

1. The Critical Early Tenure Risk Window: Data shows a critical tenure risk window early in the employee lifecycle. The average tenure of employees who left the company is exactly 5.13 years, compared to 7.30 years for those who remained. Attrition peaks sharply in the early years, indicating that onboarding and early-career engagement are vital for retention.

2. The Impact of Poor Work-Life Balance: Poor work-life balance is a primary driver of departures. Employees rating their Work-Life Balance as 1 (Poor) exhibit an exceptionally high attrition rate of 31.25%. This is more than double the attrition rate of employees who rated their balance as 3 (Good), which sits at just 14.22%.

3. Income-Driven Flight Risk: Compensation heavily correlates with retention. The median monthly income for employees who left the company is significantly lower at $3,202.00, compared to a median of $5,204.00 for retained employees. This $2,002.00 monthly deficit concentrates attrition risk heavily among lower-earning staff.

4. Role-Specific Turnover Concentrations: Attrition is highly concentrated in specific front-line and field positions. Sales Representatives experience the absolute highest turnover rate in the dataset at 39.76%, followed closely by Laboratory Technicians at 23.94% and Human Resources at 23.08%. This contrasts sharply with Research Directors, who maintain a near-zero attrition rate of 2.50%.

Task 4 — Model Building & Comparison

```python
from sklearn.model_selection import train_test_split as TTS
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.ensemble import GradientBoostingClassifier 



```

```python
X_train, X_test, y_train, y_test = TTS(data.drop('Attrition', axis=1), data['Attrition'], test_size=0.2, random_state=42)
```

```python
X_train.columns
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
continuous_cols = [
    'Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome', 
    'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears', 
    'TrainingTimesLastYear', 'YearsAtCompany', 'YearsInCurrentRole', 
    'YearsSinceLastPromotion', 'YearsWithCurrManager'
]



# 2. Fit and transform just those columns, and overwrite them in place
X_train[continuous_cols] = scaler.fit_transform(X_train[continuous_cols])
```

```python
print(type(X_train))
print(type(y_train))
# Try fitting using the underlying NumPy arrays
```

```python

# 1. Re-initialize the model to make sure LR is actually a Logistic Regression object
LR = LogisticRegression(class_weight='balanced', max_iter=1000)

# 2. Fit using the original DataFrames (no need for .values now)
LR.fit(X_train, y_train)
```

```python
y_test = pd.DataFrame(y_test)
```

```python
LogisticRegression_prediction = LR.predict(X_test)
```

```python


# Initialize Random Forest with balanced class weights
RF = RandomForestClassifier(class_weight='balanced', random_state=42)

# Train the model
RF.fit(X_train, y_train)
```

```python
Random_Forest_prediction = RF.predict(X_test)
```

```python


# print("Random Forest Confusion Matrix:")
# print(confusion_matrix(y_test, Random_Forest_prediction))

# print("\nRandom Forest Classification Report:")
# print(classification_report(y_test, Random_Forest_prediction ))
```

```python


# Initialize Gradient Boosting (Notice: No class_weight parameter available here)
GB = GradientBoostingClassifier(random_state=42)

# Train the model
GB.fit(X_train, y_train)
```

```python

GradientBoostingClassifier_prediction = GB.predict(X_test)
```

Task 7 — HR Insights & Business Recommendations




Here is the point-by-point breakdown of the model's findings and my recommendations for our retention strategy, written entirely in non-technical language.

1. Which 3 factors most strongly predict that an employee will leave?

Based on the patterns learned by our model, the three strongest indicators that an employee is at risk of leaving are Monthly Income, Age, and Total Working Years. Employees who are younger, early in their careers, or sitting at the lower end of our pay scales show the highest statistical probability of resigning.

2. Which department or job role should HR prioritize for retention efforts?

We should immediately focus our efforts on the Sales and Research & Development (R&D) departments. Looking closer at specific roles within those areas, Sales Executives and Research Scientists represent our highest volume of departures and should be our top priority for targeted retention.

3. Does salary alone explain attrition or are there other stronger factors?

No, salary is only part of the story. While baseline compensation matters significantly, lifestyle and workload pressures—specifically heavy overtime demands and a documented lack of work-life balance—are actually much stronger catalysts pushing employees to exit, even when they are satisfied with their pay.

4. 2 Concrete HR Recommendations

Proactive Milestone Conversations: HR should flag individuals in our high-risk roles (Sales Executives and Research Scientists) who are approaching critical 2-to-5-year tenure marks. We can implement proactive "stay interviews" and partner them with senior leaders in structured mentorship tracks to explicitly map out their future internal career progression.

Overtime Burnout Triggers: We should set up an automated alert system that flags when an employee's logged overtime hours cross a set threshold in a single quarter. For teams with chronically high workloads (like Sales and R&D), we can introduce flexible hybrid scheduling or a compensatory time-off policy to directly combat burnout.

5. What limitation does this model have that an HR team should be aware of?

The biggest limitation to keep in mind is that this model is a backward-looking tool built entirely on historical, numerical data. It is completely blind to real-time, qualitative human factors—like a sudden personality conflict with a new manager, personal family emergencies, or unexpected shifts in the job market. It should be used as a helpful guide to highlight risk, but it cannot replace human intuition and direct communication.
